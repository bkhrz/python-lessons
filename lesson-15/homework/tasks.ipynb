```python
import matplotlib.pyplot as plt
import numpy as np
```

Task 1

```python
x = np.linspace(-10, 10, 1000)
y = x**2 - 4*x + 4

plt.plot(x, y, label=r'$f(x) = x^2 - 4x + 4$', color='b')
plt.title('Graph')
plt.xlabel('X')
plt.ylabel('Y', rotation = 0)
plt.legend()
plt.show()
```

Task 2

```python
x = np.linspace(0, 2*np.pi, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label = r'$ \sin(x) $', color = 'r')
plt.plot(x, y2, label = r'$ \cos(x) $', color = 'k')
plt.legend()
plt.show()
```

Task 3

```python
x = np.linspace(-10, 10, 1000)

top_left = x**3
top_right = np.sin(x)
bottom_left = np.e**x
bottom_right = np.where(x>-1, np.log(x+1), None)

plt.subplot(2, 2, 1)
plt.plot(x, top_left, label = '$ f(x) = x^3 $', color = 'b')
plt.legend()
plt.subplot(2, 2, 2)
plt.plot(x, top_right, label = '$ f(x) = x^3 $', color = 'k')
plt.legend()
plt.subplot(2, 2, 3)
plt.plot(x, bottom_left, label = '$ f(x) = e^x $', color = 'r')
plt.legend()
plt.subplot(2, 2, 4)
plt.plot(x, bottom_right, label = r'$ f(x) = \log(x+1) $ (for $ x \geq 0 $', color = 'y')
plt.legend()
```

Task 4

```python
import random

x_rand = np.random.uniform(0, 10, 100)
y_rand = np.random.uniform(0, 10, 100)
markers = ['o', 's', 'D', 'v', '^', '<', '>', 'p', '*', 'h']

for i in range(0, 100):
    plt.scatter(x_rand[i], y_rand[i], marker = random.choice(markers))

plt.title('Scatter Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```

Task 5

```python
values = np.random.normal(loc = 0, scale = 1, size = 1000)

plt.hist(values, bins = 30, color = 'yellow', edgecolor = 'r', alpha = 0.6)
plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

Task 6

```python
x = np.linspace(-5, 5, 1000)
y = np.linspace(-5, 5, 1000)

xx, yy = np.meshgrid(x, y)
f = np.cos(xx**2 + yy**2)

ax = plt.axes(projection='3d')
ax.plot_surface(xx, yy, f, cmap = 'spring')
plt.title(r'$ f(x, y) = \cos(x^2 + y^2) $')
plt.show()
```

```python
periods = ['T1', 'T2', 'T3', 'T4']
a_value = [3, 5, 7, 6]
b_value = [4, 6, 2, 5]
c_value = [2, 3, 5, 4]

plt.bar(periods, a_value, label = 'Category A')
plt.bar(periods, b_value, label = 'Category B')
plt.bar(periods, c_value, label = 'Category C')

plt.title('Stacked Bar chart')
plt.ylabel('Values')
plt.xlabel('Time Periods')
plt.legend()
plt.show()
```
